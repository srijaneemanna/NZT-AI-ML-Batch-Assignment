```python
# Task 1: Python Basics & Operators
import random
import numpy as np
from datetime import datetime

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print(f"Addition: {num1 + num2}")
print(f"Subtraction: {num1 - num2}")
print(f"Multiplication: {num1 * num2}")
print(f"Division: {num1 / num2}")
print(f"Modulus: {num1 % num2}")
print(f"Power: {num1 ** num2}")

print(f"{num1} > {num2}: {num1 > num2}")
print(f"{num1} < {num2}: {num1 < num2}")
print(f"{num1} == {num2}: {num1 == num2}")
print(f"{num1} != {num2}: {num1 != num2}")
```

```python
# Task 2: Lists and Arrays
import random
import numpy as np
from datetime import datetime

numbers = [random.randint(1, 100) for _ in range(10)]
numbers.append(50)
numbers.pop(0)

print("List:", numbers)
print("Max:", max(numbers))
print("Min:", min(numbers))

array = np.array(numbers)
print("Mean:", np.mean(array))
print("Median:", np.median(array))
print("Std Dev:", np.std(array))
```

```python
# Task 3: Dictionary & Sets

import random
import numpy as np
from datetime import datetime

student = {"name": "Rahul", "course": "AI/ML", "marks": 82}

if student["marks"] >= 80:
    student["grade"] = "A"
elif student["marks"] >= 60:
    student["grade"] = "B"
else:
    student["grade"] = "C"

for k, v in student.items():
    print(k, ":", v)

ai_tools_1 = {"ChatGPT", "TensorFlow", "PyTorch"}
ai_tools_2 = {"PyTorch", "Keras", "Scikit-learn"}
print("Common AI Tools:", ai_tools_1 & ai_tools_2)
```
